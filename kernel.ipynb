```python
import numpy as np
from scipy.misc import imresize
from sklearn.cluster import k_means
import matplotlib.pyplot as plt
from imageio import imread, imsave
%matplotlib inline
```

```python
k = 7     #no. pf colors
```

```python
img = imread('image.jpg')
img.shape
```

```python
plt.imshow(img)
```

```python
def resize_dims(dims):
    n = max(dims)
    divider = np.ceil(n/500)
    dim0 = int(dims[0]//divider)
    dim1 = int(dims[1]//divider)
    return (dim0, dim1)
```

```python
img_new = imresize(img, resize_dims(img.shape))
plt.imshow(img_new)
plt.show()
img_new = np.reshape(img_new, (img_new.shape[0]*img_new.shape[1],3))
img_new.shape
```

```python
centroids, idx, _ = k_means(img_new, n_clusters=10)
```

```python
compressed = centroids[list(map(int, idx))]
compressed = np.reshape(compressed,(*resize_dims(img.shape), 3))
```

```python
imsave("new_img.jpg", compressed)
```
